```python
def filtro_umbralizacion(img):
    hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

    bajo1 = np.array([15, 70, 70])   
    alto1 = np.array([10, 255, 255])

    bajo2 = np.array([160, 30, 30])
    alto2 = np.array([180, 255, 255])

    mask1 = cv2.inRange(hsv, bajo1, alto1)
    mask2 = cv2.inRange(hsv, bajo2, alto2)
    mask_final = cv2.bitwise_or(mask1, mask2)

    resultado = cv2.bitwise_and(img, img, mask=mask_final)
    return resultado
```

```python
import os
import cv2
import numpy as np

imagenes_train = "dataset/train/images"
imagenes_valid = "dataset/valid/images"
imagenes_test = "dataset/test/images"

for fichero in [imagenes_train, imagenes_valid, imagenes_test]:
    for i in os.listdir(fichero):
        path_completo = os.path.join(fichero, i)
        img = cv2.imread(path_completo)
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        resultado = filtro_umbralizacion(img_rgb)
        resultado = cv2.resize(resultado, (256, 256), interpolation=cv2.INTER_AREA)
        img_para_guardar = cv2.cvtColor(resultado, cv2.COLOR_RGB2BGR)
        ruta_guardado = os.path.join(fichero, i)
        exito = cv2.imwrite(ruta_guardado, img_para_guardar)
            
```

```python
model = YOLO("yolo11n.pt")
results = model.train(
    data="dataset/config.yaml", 
    epochs=100, 
    imgsz=256, 
    device=0
)
```
